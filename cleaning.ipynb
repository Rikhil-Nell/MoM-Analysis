```python
import pandas as pd
```

```python
df= pd.read_csv('./mom_data.csv')
```

```python
df = df.drop(columns=['discount','total_tax','sap_code','delivery_charge', 'container_charge', 'service_charge','additional_charge', 'waived_off', 'round_off','customer_address', 'persons','order_cancel_reason','invoice_no','payment_type', 'order_type','status', 'area', 'virtual_brand_name', 'brand_grouping', 'assign_to',])
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d %H:%M:%S').dt.strftime('%d-%m-%Y').where(df['total']!=0)
# df = df.where(df['total']!=0)
```

```python
# Visit frequency
df['day'] = pd.to_datetime(df['date'], format='%d-%m-%Y').dt.day_name()
df['month-year'] = pd.to_datetime(df['date'], format='%d-%m-%Y').dt.strftime('%Y-%b')
day_wise_peak_visit_times = df.groupby('day')['customer_phone'].count().reset_index(name='day wise visit count').sort_values(by='day wise visit count', ascending=False)
monthly_peak_visit_times = df.groupby('month-year')['customer_phone'].count().reset_index(name='monthly visit count').sort_values(by='monthly visit count', ascending=False)

# Customer Lifetime Value (CLV)
clv = df.groupby('customer_phone',)['total'].sum().reset_index(name='CLV').sort_values(by='CLV', ascending=False)

# Average Order Value (AOV)
aov = df.groupby('customer_phone')['total'].mean().reset_index(name='AOV').sort_values(by='AOV', ascending=False)

# Average Spend per Visit
average_spend_per_visit = df.groupby('customer_phone')['total'].mean().reset_index(name='Average Spend per Visit').sort_values(by='Average Spend per Visit', ascending=False)
```

```python
average_spend_per_visit
```

```python
# Top-Selling Items
top_selling_items = df.groupby('item_name')['item_quantity'].sum().reset_index().sort_values(by='item_quantity', ascending=False)
```

```python
latest_date = df['date'].max()
recency = df.groupby('customer_phone')['date'].max().reset_index()
recency['latest_date'] = latest_date
recency['recency'] = pd.to_datetime(recency['latest_date'], format='%d-%m-%Y') - pd.to_datetime(recency['date'], format='%d-%m-%Y')
```

```python
# Recency of Visit
latest_date = df['date'].max()
recency = df.groupby('customer_phone')['date'].max().reset_index()
recency['Recency'] = (latest_date - recency['date']).dt.days()
recency = recency[['customer_phone', 'Recency']]
```
